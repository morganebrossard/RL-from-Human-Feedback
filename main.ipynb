```python
import torch
from ac_mlp import ActorMLP, CriticMLP
from rl import ReinforcementLearningFromHumanPreferences
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device: ", device)
```

```python
model = ReinforcementLearningFromHumanPreferences(
    state_shape=(8, 84, 84),
    random_seed=0,
    compute_device=device,
    discount_factor=0.99,
    actor_lr=0.001,
    critic_lr=0.005,
    update_frequency=10,
    actor_network=ActorMLP,
    critic_network=CriticMLP,
    use_real_human=True,
)
```

```python
scores = model.train_from_human_preferences(episode_count=10)
```
